# **Doctor Name**

```python
doctor_name_tag=soup.find_all("td",{"data-title":"Name"}) ### Parse out by right clicking Inspect and go through the Html Code on Web page link  
doctors_name_1=[]

for i in doctor_name_tag:
    doctors_name_1.append(i.text.strip())
print(doctors_name_1)
```

```python
spec_list=[]
spec_tag= soup.find_all("td",{"data-title":"Special."})
#len(spec_tag)
for i in spec_tag:
    spec_list.append(i.text.strip())
print(spec_list)
```

```python
degree_list =[]
degree_tag= soup.find_all("td",{"data-title":"Degree"})
for i in degree_tag:
    degree_list.append(i.text.strip())
print(degree_list)
```

# **Extracting Contact Number using details link**

```python
## USING re library for extracting phone number 
import re
```

```python
# using for loop to extract conatct numbers by using details link 
contact_number=[]
# create a contact number empty list to append and stroe the numbers 

for link in detail_link:
    r1=requests.get(link)
#using try and except to find out which doctor phone number is available aon web page and which not 
    try:
        "Phone Number" in r1.text # search out phone number in r1.text
        match = re.search("Phone Number</strong></td> <td>(\d+)-(\d+)",r1.text)
        a=str(match.group(1))
        b=str(match.group(2))
        contact_number.append(a+b) # append the number as a string 
        print(a+b)
    except:
        "Phone Number" not in r1.text # if no not availble on web page append nil which can be dealt in later stages
        print("nil")
        contact_number.append("nil")
```

```python
dict_1= {"contact nunber" :contact_number}
```

# **Creating Dataframe With Multiple Pages details** 

1.Next Target is extracting out all the data through multiple pages by using Function Loops 

2.Now, the Nomeclature of Name||Specilization||Degree||State||Links||Numbers might get changed 

```python
#Creating Final list for Doctors info
doctorlist=[]

# using for loop to iterate multiple webpages availble on website 
for page in range(1,800,5):
    
    doc_url="https://www.drdata.in/list-doctors.php?search=Doctor&page="
    # use str(page) toiterate through multipkle pages and loop the data againa nd again 
    response=requests.get(doc_url + str(page))
    soup=BeautifulSoup(response.text,"html.parser")
    
    #container is nothing just a variable which store all the table availble on webpages
    container =soup.find("table",{"class":"table-bordered table-striped table-condensed cf"})
    # tr is nothing just rows ( abbreviation used in html language)
    rows= container.find_all("tr")
    
    for row in rows[1:16]:
        dictionary= {
            "doctor" : row.find("td",{"data-title":"Name"}).text.strip() , # using strip to extract the exact data
            #print(doctor)
            "specialization" : row.find("td",{"data-title":"Special."}).text.strip() ,
            #print(specialization)
            "degree" : row.find("td",{"data-title":"Degree"}).text.strip() ,
            #print(degree)
            "state" : row.find("td",{"data-title":"State"}).text.strip() ,
            #print(state)
            "details" : "http://www.drdata.in/" + row.find("a")["href"] ,
            #print(details)
        }
        
        ## finally append the dictionary in a list
        doctorlist.append(dictionary)
```

# **Multiple Page contact Numbers**

```python
#create a new empty list for all links 
links=[]

for i, rows in doctor_df.iterrows(): ### iterate through rowsd using iterrows function 
    links.append(rows[4])
```

```python
links[1:10]
```

```python
# create a new empty list for phone numbers of all doctors
Phone_Number=[]

for link in links:
    r=requests.get(link)
    
    try:
        "Phone Number" in r.text
        match = re.search("Phone Number</strong></td> <td>(\d+)-(\d+)",r.text)
        a=str(match.group(1))
        b=str(match.group(2))
        Phone_Number.append(a+b)
        print(a+b)
    except:
        "Phone Number" not in r.text
        print("nil")
        Phone_Number.append("nil")
```

```python
len(Phone_Number)
```

```python
dict_2= {"Phone Number": Phone_Number}
```
